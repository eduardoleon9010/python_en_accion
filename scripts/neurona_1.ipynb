# Cálculo de la salida de una neurona artificial con función de activación sigmoide

Este código implementa una neurona artificial con dos entradas y una salida,
utilizando la función de activación sigmoide.

red neuronal.png

## Parámetros de la neurona
- **b (bias):** 0.15
- **w1 (peso de la primera entrada):** 0.55
- **w2 (peso de la segunda entrada):** 0.45
- **x1 (primera entrada):** 0.5
- **x2 (segunda entrada):** -0.35

La salida de la neurona se calcula en dos pasos:
1. **Cálculo de la combinación lineal:**
   [ z = b + (w_1 \ x_1) + (w_2 \ x_2)]
2. **Aplicación de la función de activación sigmoide:**
   [sigma(z) = \{1}{1 + e^{-z}}]

El resultado final se muestra con tres decimales de precisión.

```python
import math

# Definición de parámetros
b = 0.15
w1 = 0.55
w2 = 0.45
x1 = 0.5
x2 = -0.35

# Cálculo de la combinación lineal
z = b + (w1 * x1) + (w2 * x2)

# Aplicación de la función sigmoide
sigmoid = 1 / (1 + math.exp(-z))

# Resultados con tres decimales
z, round(sigmoid, 3)
```
